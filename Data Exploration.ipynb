```python
# Read in the dataset
import csv
import pandas as pd
from tableone import TableOne

data = pd.read_csv("data/covars_and_biological_ds5500.csv")
for column in data.columns:
    print(column)


# From here we are going to generate a Table 1
columns = ['ISAGE', 'FVCURRWT', 'PRETERM_BEST', 'CHILDNUM', 'PREGNUM']
categorical = ['PRETERM_BEST']
group_by  = 'PRETERM_BEST'
mytable = TableOne(data, columns, categorical, group_by, pval = True)
mytable.to_latex("latextable")
```
